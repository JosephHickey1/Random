```python
import numpy as np
import astropy.units as u
import astropy.constants as const
import matplotlib.pyplot as plt
from IPython.display import Latex
```

# Forward Euler method of integration using a constant position change, dx
def Integrator(psi,dpsi,x,dx,E):
    # psi(x+dx) = psi(x) + dx*dpsi/dx|x + 0.5*dx**2 *d2psi/dx2|x
    # dpsi/dx|x+dx = dspi/dx|x + dx*d2psi/dx2|x
    # d2psi/dx2|x = 2m/hbar * (V(x) - E) * psi
    # psi(0) = 0
    hbar = const.hbar.to(u.eV*u.s).value
    m_e = const.m_e.to(u.eV/(const.c.to(u.AA/u.s))**2).value
    
    d2psi = 2*m_e/(hbar**2) * psi * (Potential(x) - E)
    dpsi += (dx * d2psi)
    psi += (dx*dpsi + 0.5*(dx**2)*d2psi)
    
    return psi,dpsi


```python
def Integrator(psi,dpsi,x,dx,E):
    # psi(x+dx) = psi(x) + dx*dpsi/dx|x + 0.5*dx**2 *d2psi/dx2|x
    # dpsi/dx|x+dx = dspi/dx|x + dx*d2psi/dx2|x
    # d2psi/dx2|x = 2m/hbar * (V(x) - E) * psi
    # psi(0) = 1 since half potentials do not conform to the odd states of
    # the symmetric potentials
    
    d2psi = 0.26426 * psi * (Potential(x) - E)
    psi += (dx*dpsi + 0.5*(dx**2)*d2psi)
    dpsi += (dx * d2psi)
    
    return psi,dpsi
```

```python
def tracker(sign,xf,dx,E):
    
    n = int(np.ceil(xf/dx))
    psi = np.zeros(n)
    dpsi = np.zeros(n)
    d2psi = np.zeros(n)
    
    if sign == 'odd':
        psi[0] = 1
    elif sign == 'even':
        dpsi[0] = 1
    
    for i in range(n-1):
        
        psi[i+1],dpsi[i+1] = Integrator(psi[i],dpsi[i],i*dx,dx,E)
        
    
    x = np.arange(0,xf,dx)
    
    return x,psi
        
    
    
    
```

```python
def Potential(x):
    if x <= 0:
        x = 100000
    
    alpha = 1
    
    return alpha*x
```

```python
x,p = tracker('odd',15,0.001,3.60594058)
plt.plot(x,p,label='Test Wavefunction Psi')
y=np.zeros(len(x))
plt.plot(x,y,'k--',label='Psi=0')
plt.legend(loc='upper right')
plt.title('Wavefunction with ' r'$E_1 = 3.60594058$')
plt.xlabel('Position ' r'($\AA$)')
```
